```python
from classy import Class
```

```python
font = {'size'   : 22, 'family':'STIXGeneral'}
axislabelfontsize='x-large'
matplotlib.rc('font', **font)
matplotlib.mathtext.rcParams['legend.fontsize']='medium'
```

```python
LCDM = Class()
LCDM.set({'Omega_cdm':0.25,'Omega_b':0.05})
LCDM.compute()
LCDM.cleanup()
```

```python
CDM = Class()
CDM.set({'Omega_cdm':0.95})
CDM.compute()
CDM.cleanup()

Omega_lambda = CDM.get_current_derived_parameters(['Omega0_lambda'])
print Omega_lambda
```

```python
baLCDM = LCDM.get_background()
baCDM = CDM.get_background()
baCDM.viewkeys()
```

```python
fLCDM = LCDM.Hubble(0)
fCDM = CDM.Hubble(0)
```

```python
namelist = ['lum. dist.','comov. dist.','ang.diam.dist.']
colours = ['b','g','r']
for name in namelist:
    idx = namelist.index(name)
    loglog(baLCDM['z'],fLCDM*baLCDM[name],colours[idx]+'-')
legend(namelist,loc='upper left')
for name in namelist:
    idx = namelist.index(name)
    loglog(baCDM['z'],fCDM*baCDM[name],colours[idx]+'--')
xlim([0.07, 10])
ylim([0.08, 20])
xlabel(r"$z$")
ylabel(r"$\mathrm{Distance}\times H_0$")
```
